### Description of the dataset

This dataset contains statistics on the world's billionaires, including information about their businesses, industries, and personal details. It provides insights into the wealth distribution, business sectors, and demographics of billionaires worldwide.

source: https://www.kaggle.com/datasets/nelgiriyewithana/billionaires-statistics-dataset

- **rank** — ranking of the billionaire in terms of wealth
- **finalWorth** — final net worth in U.S. dollars
- **category** — industry or category of the billionaire's business
- **personName** — full name of the billionaire
- **age** — age of the billionaire
- **country** — country of residence
- **city** — city of residence
- **source** — source of the billionaire's wealth
- **industries** — industries associated with the billionaire's business interests
- **countryOfCitizenship** — country of citizenship
- **organization** — associated organization or company
- **selfMade** — whether the billionaire is self-made (True/False)
- **status** — "D" = self-made, "U" = inherited/unearned wealth
- **gender** — gender of the billionaire
- **birthDate** — birthdate
- **lastName** — last name
- **firstName** — first name
- **title** — title or honorific
- **date** — date of data collection
- **state** — state of residence
- **residenceStateRegion** — region or state of residence
- **birthYear** — birth year
- **birthMonth** — birth month
- **birthDay** — birth day
- **cpi_country** — Consumer Price Index (CPI) for the billionaire's country
- **cpi_change_country** — CPI change for that country
- **gdp_country** — Gross Domestic Product for that country
- **gross_tertiary_education_enrollment** — tertiary education enrollment rate
- **gross_primary_education_enrollment_country** — primary education enrollment rate
- **life_expectancy_country** — life expectancy in the country
- **tax_revenue_country_country** — tax revenue for the country
- **total_tax_rate_country** — total tax rate for the country
- **population_country** — population of the country
- **latitude_country** — latitude coordinate of the country
- **longitude_country** — longitude coordinate of the country
